### Import Artist Genres Dataframe

```python
df_artist_genres = pd.read_csv('../data/df_artist_genres_pop.csv')
```

```python
df_artist_genres = df_artist_genres.drop(['Unnamed: 0','track_name','artist_id','artist_name'],
                                        axis = 1,
                                        )
df_artist_genres.head()
```

## Create Modeling Data Set

```python
cols = ['track_id', 'track_popularity',
       'track_duration', 'artist_popularity', 'artist_follower_total',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
df_final = df[cols]
df_final.shape
```

```python
df_final = pd.merge(left = df_final,
                   right = df_artist_genres,
                   how = 'inner',
                   on = 'track_id')
df_final.shape
```

```python
df_final.head()
```

### Save Final Modeling Dataset

```python
df_final.to_csv('../data/df_final.csv')
```
